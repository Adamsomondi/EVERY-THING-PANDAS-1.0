```python
#remember when we created series from dictionaries you can also do that here.
#fun right and create even a more complex dictionaries with arrays inside.

import pandas as pd
city_data = {
    "City": ["New York City", "Paris", "Barcelona", "Rome"],
    
    "Country": ["United States", "France", "Spain", "Italy"],
    
    "Population": pd.Series([8600000, 2141000, 5515000, 2873000])
}

cities = pd.DataFrame(city_data)
cities
```

```python
# transposing in linear algebra class made easy ha ha ha ha ha 
cities.transpose()
```

```python
import pandas as pd
import numpy as np
random_data = np.random.randint(1, 101, [3, 5])
random_data
```

```python
pd.DataFrame(data = random_data)
```

```python
#more fun right more control but still simple you may actually work with data frames alone
import numpy as np
row_labels = ["Morning", "Afternoon", "Evening"]
random_data=np.random.randint(1,101,[3,5])
temperatures = pd.DataFrame(  data = random_data, index = row_labels)
temperatures
```

```python
row_labels = ["Morning", "Afternoon", "Evening"]
column_labels = (
    "Monday", "Tuesday", "Wednesday", "Thursday", "Friday")
pd.DataFrame(
    data = random_data,
    index = row_labels,
    columns = column_labels,
)
```

```python
row_labels = ["Morning", "Afternoon", "Morning"]
column_labels = [    "Monday", "Tuesday", "Wednesday","Tuesday", "Friday"]
pd.DataFrame(
    data = random_data,
    index = row_labels,
    columns = column_labels,
)
```

## I promised i will show you the great  Sorting  algorithm lets learn it in  a DataFrame

```python
##i want to sort only a single column

import pandas as pd
nba=pd.read_csv("nba.csv")
nba.sort_values("Name")
```

```python
#pretty arguments right i sort from the bottom up they are just arguments don't panick
nba.sort_values("Name", ascending = True).tail()
```

```python
nba.sort_values("Birthday", ascending = False).head()
```

```python
#okay i want to sort two columns u see more controls
nba.sort_values(by = ["Team", "Name"])
```

```python
nba.sort_values(["Team", "Name"], ascending = False)
```

```python
nba.sort_values(
    by = ["Team", "Salary"], ascending = [True, False]
)
```

```python
nba = nba.sort_values(
    by = ["Team", "Salary"],
    ascending = [True, False]
)
```
